```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

```python
dogs_and_cats = pd.read_csv('dogs_n_cats.csv')
dogs_and_cats
```

```python
dogs_and_cats.isna().sum()
```

```python
y = dogs_and_cats['Вид']
y
```

```python
X = dogs_and_cats.drop('Вид', axis = 1)
```

```python
X
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
X_train, X_test, y_train, t_test = train_test_split(X, y, test_size = 0.33, random_state = 43)
```

```python
clf.fit(X_train, y_train)
clf.score(X_train, y_train)
```

```python
df_ts = pd.read_json('dataset_209691_15.txt')
df_ts
```

```python
res = clf.predict(df_ts)
pd.Series(res)[res == 'собачка'].count()
```
